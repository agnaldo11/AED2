1. (ENADE) Grafo e definições. Considere as seguintes afirmações sobre grafos não dirigidos:

(A) Pseudografo: grafo que pode ter arestas paralelas e laços.
(B) Multigrafo: grafo que pode ter arestas paralelas, mas não tem laços.
(C) Grafo completo: grafo simples no qual todo par de vértices está conectado por uma aresta.
(D) Grafo nulo: grafo sem nenhuma aresta (vértices isolados).

Sobre as afirmações, assinale a alternativa correta:

(a) Apenas A e C estão corretas.
(b) Apenas B e D estão corretas.
(c) Apenas A, B e C estão corretas.
(d) Apenas C e D estão corretas.
(e) Todas estão corretas. (x)

Resposta: (e).

A afirmação (A) define corretamente pseudografo; (B) define multigrafo sem laços; (C) define grafo completo; (D) define grafo nulo, que está correto. Portanto, todas estão corretas.

2. (Busca em largura – BFS) Sobre o algoritmo de busca em largura (BFS), é correto afirmar:

(a) O BFS utiliza uma pilha para visitar vértices.
(b) O BFS explora recursivamente até não encontrar mais vértices.
(c) O BFS visita os vértices por ordem de distância mínima (níveis) a partir da origem (x)
(d) O BFS garante encontrar o ciclo mais longo num grafo.
(e) O BFS só funciona em grafos não dirigidos.

Resposta: (c).

Em BFS, os vértices são visitados em ordem de distância crescente (níveis) a partir do vértice inicial. A alternativa (a) está incorreta porque BFS usa fila, não pilha (pilha seria DFS). A alternativa (b) descreve DFS (busca recursiva em profundidade), não BFS. (d) é incorreto (BFS não busca ciclos, e certamente não maximiza); (e) é falso, BFS funciona também em grafos dirigidos. Portanto, (c) é a única correta.

3. Implementar DFS em Python: Dado o grafo abaixo representado por lista de adjacências, escreva uma função `dfs(grafo, inicio)` que retorna a lista de vértices visitados em ordem de descoberta pela busca em profundidade. Mostre a execução partindo de 'A'.

```python
# Grafo de exemplo

G = {
'A': ['B', 'C'],
'B': ['A', 'D', 'E'],
'C': ['A', 'F'],
'D': ['B'],
'E': ['B', 'F'],
'F': ['C', 'E']
}

# Função DFS recursiva

def dfs(grafo, inicio, visitados=None):
    if visitados is None:
        visitados = []
    visitados.append(inicio)
    for viz in grafo[inicio]:
        if viz not in visitados:
            dfs(grafo, viz, visitados)
    return visitados

# Execução

ordem = dfs(G, 'A')
print(ordem)
```

Solução comentada:A função `dfs` inicia adicionando o vértice atual à lista de visitados. Para cada vizinho não visitado, chama recursivamente dfs.

No grafo dado, partindo de 'A', a ordem de visita pode ser, por exemplo, `['A', 'B', 'D', 'E', 'F', 'C']`.

Isso ocorre porque o DFS vai de A a B, depois de B a D (sem mais opções), volta para B e vai a E, de E a F, de F a C (como C é vizinho de F e não visitado ainda), e então termina.

A saída específica pode variar conforme a ordem das listas de adjacência, mas seguirá a lógica profunda.

4. Implementar BFS em Python: Dado o mesmo grafo acima, escreva uma função bfs(grafo, inicio) que retorna a lista de vértices em ordem de descoberta pela busca em largura. Execute a partir de 'A'.

```python
from collections import import deque
def bfs(grafo, inicio):
    visitados = [inicio]
    fila = deque([inicio])
    ordem = []
    while fila:
        v = fila.popleft()
        ordem.append(v)
        for viz in grafo[v]:
            if viz not in visitados:
                visitados.append(viz)
                fila.append(viz)
    return ordem
        
ordem_bfs = bfs(G, 'A')
print(ordem_bfs)
```

```python
Solução comentada: A função bfs usa uma fila para visitar vértices por níveis. Inicialmente, fila = ['A']. Em cada iteração, retira o próximo vértice v, adiciona-o à ordem de saída e insere todos os vizinhos não visitados na fila. Para o grafo dado, a saída será ['A', 'B', 'C', 'D', 'E','F']: primeiro A, depois todos os vizinhos de A (B e C), depois os vizinhos de B e C (D, E e F, nesta ordem). Observa-se que todos os vértices a distância 1 de A vêm antes de quaisquer vértices a distância 2.
```
