EXERCÍCIOS ALGORÍTIMO DE ORDENAÇÃO 

Vamos praticar o entendimento dos algoritmos de ordenação com alguns exercícios e problemas, incluindo questões de provas (ENADE) e questões adicionais. As soluções comentadas são apresentadas em seguida de cada enunciado.

1 (ENADE) Julgue os itens a seguir, acerca de algoritmos para ordenação:

I. O algoritmo de ordenação por inserção (Insertion Sort) tem complexidade O(n \times \log n).

II. Um algoritmo de ordenação é dito estável caso ele não altere a posição relativa de elementos de mesmo valor.

III. No algoritmo Quicksort, a escolha do elemento pivô influencia o desempenho do algoritmo.

IV. O Bubble Sort e o algoritmo de ordenação por inserção fazem, em média, o mesmo número de comparações.

Estão certos apenas os itens:

A) I e II.

B) I e III.

C) II e IV.

D) I, III e IV. (X)

Resolução: Vamos analisar cada afirmativa:

Item I: FALSO. A complexidade do algoritmo de ordenação por inserção (Insertion Sort) é O(n^2) no pior caso, e não O(n \log n). O valor O(n \log n) corresponde à complexidade de algoritmos mais eficientes, como Merge Sort ou Quick Sort (no caso médio e melhor caso). Portanto, a afirmação I está incorreta.

Item II: VERDADEIRO. Um algoritmo de ordenação é dito estável se ele preserva a ordem relativa de elementos com valores iguais. Ou seja, se dois elementos A e B têm chaves iguais e A vem antes de B na lista original, em uma ordenação estável A continuará vindo antes de B na lista ordenada. Essa é exatamente a definição apresentada, então o item II está correto.

Item III: VERDADEIRO. No algoritmo Quicksort, a escolha do pivô é fundamental para o desempenho. Um pivô mal escolhido (por exemplo, sempre escolhendo o menor ou o maior elemento quando a lista já está ordenada, ou um elemento extremo em listas não favoráveis) pode levar o algoritmo a dividir os subproblemas de forma muito desigual, resultando em tempo de execução O(n^2) no pior caso. Por outro lado, uma boa escolha de pivô tende a equilibrar as divisões, mantendo o desempenho médio em O(n \log n).

Item IV: VERDADEIRO. Tanto o Bubble Sort quanto o Insertion Sort realizam, em média, um número semelhante de comparações, ambos de ordem O(n^2). Embora na prática o Insertion Sort tenda a ser um pouco mais eficiente (especialmente em listas quase ordenadas), do ponto de vista de complexidade assintótica média, ambos estão na ordem quadrática e realizam comparações em quantidade da mesma ordem de grandeza. Assim, a afirmativa IV está correta.

Agora, identificando quais itens estão certos: II, III e IV são verdadeiros, enquanto I é falso. Isso corresponde à alternativa E) II, III e IV.


2 (ENADE) Considere o algoritmo que implementa o seguinte processo: uma coleção desordenada de elementos é dividida em duas metades e cada metade é utilizada como argumento para a reaplicação recursiva do procedimento. Os resultados das duas reaplicações são, então, combinados pela intercalação dos elementos de ambas, resultando em uma coleção ordenada. Qual é a complexidade desse algoritmo?

Alternativas:

A) O(n^2)

B) O(n^{2n})

C) O(2^n)

D) O((\log n) \times (\log n))

E) O(n \times \log n) (X)

Resolução: O algoritmo descrito é claramente o Merge Sort, pois ele divide recursivamente a coleção em duas metades e depois intercala (merge) as partes ordenadas. Sabemos que o Merge Sort tem complexidade de tempo O(n \log n). Para justificar:

A cada nível de recursão, o conjunto é dividido em duas partes (essa divisão produz aproximadamente \log_2 n níveis de recursão, pois dividir n repetidamente por 2 até chegar a 1 produz \log_2 n divisões).

Em cada nível de recursão, ocorre o passo de intercalação (merge) dos sub-resultados, que percorre todos os elementos para combinar as listas (custo O(n) por nível, somando as intercalações das sublistas daquele nível).

Com \log n níveis e processamento de n elementos em cada nível (no total das merges daquele nível), o trabalho total é aproximadamente n \times \log n.

Matematicamente, o custo T(n) do Merge Sort satisfaz a recorrência T(n) = 2T(n/2) + O(n), cuja solução por método mestre é O(n \log n). Portanto, a resposta correta é a alternativa E) O(n \times \log n).


3 (ENADE) Considere os seguintes algoritmos de ordenação e suas respectivas características:

I. O Merge Sort é um algoritmo de ordenação estável com complexidade O(n log n) no pior caso.

II. O Quick Sort tem complexidade O(n^2) no pior caso, mas seu desempenho médio é O(n log n).

III. O Selection Sort tem complexidade O(n^2) no pior caso e não é um algoritmo estável.

IV. O Insertion Sort tem complexidade O(n^2) no pior caso, mas pode ser eficiente em listas quase ordenadas, com complexidade próxima de O(n).

Estão corretos os itens:

A) I e II

B) II e III

C) I, II e IV

D) I, III e IV (X)

E) II, III e IV

Resolução: Vamos verificar cada afirmativa:

Item I: Verdadeiro. O Merge Sort, como vimos, é estável (mantém a ordem relativa de elementos iguais) e possui complexidade O(n log n) em todos os casos, inclusive no pior caso. Portanto, I está correto.

Item II: Verdadeiro. O Quick Sort tem, de fato, um pior caso O(n^2) (quando as partições são muito desequilibradas), mas tipicamente seu desempenho médio é O(n log n). Essa é uma afirmação padrão sobre Quick Sort. Então II está correto.

Item III: Verdadeiro. O Selection Sort realiza O(n^2) comparações no pior caso (e em qualquer caso, na verdade, pois não depende da ordem dos dados) e não é estável, pois ao fazer a troca do menor elemento encontrado com o elemento inicial da seção não ordenada, dois elementos iguais poderiam inverter a ordem relativa. Assim, III está correto.

Item IV: Verdadeiro. O Insertion Sort tem complexidade O(n^2) no pior caso, mas em listas que já estão quase ordenadas (por exemplo, apenas alguns elementos fora do lugar), seu desempenho pode se aproximar de O(n). Isso ocorre porque ele precisará deslocar muito poucos elementos para inserir cada item na posição correta. Essa propriedade é conhecida: insertion sort é eficiente para entradas quase ordenadas. Portanto, IV também está correto.

Como todos os quatro itens I, II, III e IV foram avaliados como verdadeiros, a alternativa que corresponde é D) I, III e IV.


4 (ENADE) Considere dois algoritmos de ordenação: Bubble Sort e Quick Sort. Suponha um vetor de 10.000 elementos, onde todos os elementos estão em ordem decrescente (ou seja, na pior ordem possível para alguns métodos). Sobre a execução desses algoritmos nesse cenário, julgue os itens a seguir:

O Bubble Sort terá desempenho O(n^2), fazendo um grande número de comparações e trocas.

O Quick Sort pode ter desempenho O(n^2), caso o pivô seja escolhido de forma ineficiente, como sempre o primeiro ou o último elemento da lista.

O Bubble Sort é mais eficiente que o Quick Sort em listas completamente desordenadas.

O Quick Sort pode ser mais rápido que o Bubble Sort, se uma boa escolha de pivô for feita, mesmo em uma lista inicialmente em ordem decrescente.

Estão corretos os itens:

A) I e II

B) II e IV

C) I, II e IV (X)

D) II e III

E) I e III

Resolução: Avaliando cada item:

Item I: Verdadeiro. O Bubble Sort, para uma lista inversamente ordenada (decrescente quando queremos ascendente), irá realizar o máximo de operações: a cada comparação de elementos adjacentes ele vai encontrar que precisa trocar, empurrando cada elemento para a sua posição correta lentamente. Isso resulta em um número de comparações e trocas da ordem de \frac{n(n-1)}{2}, que é O(n^2). Com 10.000 elementos em ordem decrescente, o Bubble Sort fará um número muito grande de comparações e trocas, confirmando o comportamento O(n^2). Então o item I está correto.

Item II: Verdadeiro. O Quick Sort, como sabemos, pode degradar para O(n^2) se o pivô for mal escolhido em cada partição. Uma forma “ineficiente” de escolher pivô é sempre pegar, por exemplo, o último elemento. Se os dados estiverem em ordem decrescente e o algoritmo escolher sempre o último elemento como pivô, então inicialmente o último elemento será o menor de todos (porque a lista está decrescente), resultando em partições extremamente desbalanceadas (todos os elementos irão para a esquerda do pivô, e nenhum para a direita). Isso repetido recursivamente leva ao pior caso O(n^2). Portanto, a afirmação II está correta ao dizer que Quick Sort pode ter desempenho O(n^2) se o pivô for escolhido de forma desfavorável.

Item III: Falso. O Bubble Sort não é mais eficiente que o Quick Sort em listas completamente desordenadas (ou mesmo em ordem decrescente). Na verdade, Quick Sort, na média, tende a ser muito mais rápido que Bubble Sort mesmo em listas bem desordenadas. O Bubble Sort, como algoritmo O(n^2), vai ser muito mais lento que o Quick Sort que, com um pivô aleatório ou mediana, ainda conseguirá próximo de O(n log n) comparações. Portanto, o item III está incorreto. (Observação: Talvez a afirmativa III tenha tentado sugerir que no pior caso Quick Sort degrada e Bubble Sort também é O(n^2), mas mesmo assim Quick Sort típico não é pior que Bubble Sort nessa situação com boas escolhas de pivô. De qualquer forma, a frase “Bubble Sort é mais eficiente que Quick Sort em listas completamente desordenadas” não é verdadeira para entradas grandes.)

Item IV: Verdadeiro. O Quick Sort pode ser bem mais rápido que o Bubble Sort se for feita uma boa escolha de pivô, mesmo em um caso adverso como lista ordenada decrescentemente. Por exemplo, se escolhermos sempre o pivô como a mediana (ou próximo disso) dos elementos, a ordem decrescente não será um problema e Quick Sort executará perto de O(n log n). Mesmo sem escolha ideal, mas aleatória, é improvável que Quick Sort degrade completamente; na maioria dos casos terá divisões razoáveis e superará amplamente o Bubble Sort. Portanto, IV está correto.

Concluindo, os itens corretos são I, II e IV. Isso corresponde à alternativa C) I, II e IV.


5 Suponha que você tenha um vetor quase ordenado, isto é, ele já está praticamente em ordem crescente, exceto por alguns poucos elementos fora de lugar. Qual dos algoritmos de ordenação estudados tenderia a apresentar o melhor desempenho nesse caso? Justifique.

Resolução: Para uma lista que já está quase ordenada, o algoritmo que apresentará o melhor desempenho é o Insertion Sort. Isso porque o Insertion Sort aproveita muito a ordenação pré-existente: ele apenas percorre comparando e faz pouquíssimas movimentações quando os elementos já estão próximos de suas posições corretas. No melhor cenário (lista já ordenada), o Insertion Sort executa em tempo linear O(n), pois cada novo elemento inserido já é maior que todos os da esquerda e não precisa ser movido. Se apenas alguns elementos estão fora de lugar, o número de deslocamentos necessários será pequeno, mantendo o tempo de execução próximo de linear.

Em contrapartida, algoritmos como Bubble Sort e Selection Sort, mesmo com a lista quase ordenada, ainda fariam O(n^2) comparações (Bubble Sort poderia terminar mais cedo se implementado com verificação de trocas, mas ainda assim faria muitas comparações iniciais; Selection Sort não ganha nada com ordenação parcial, sempre varre todo o vetor para achar mínimos). Quick Sort e Merge Sort teriam seus usuais O(n log n) tanto faz (não pioram, mas também não melhoram muito além disso, e n log n para n grande é maior que n se a lista já está ordenada). Já o Insertion Sort se beneficia substancialmente de entradas quase ordenadas, podendo inclusive superar algoritmos O(n log n) em casos moderados de quase ordenação, pois se o vetor estiver k-ordenado (cada elemento deslocado no máximo k posições de onde deveria), o Insertion Sort tem complexidade O(n * k). Por isso, para listas quase ordenadas, Insertion Sort é a melhor escolha dentre os algoritmos apresentados.


6 Estabilidade na ordenação: Quais dos algoritmos de ordenação estudados são estáveis? Explique brevemente o que significa um algoritmo ser estável e forneça um exemplo de situação em que a estabilidade é importante.**

Resolução: Dos algoritmos que vimos: Bubble Sort, Insertion Sort e Merge Sort são estáveis (em suas implementações padrão). Selection Sort, Quick Sort e Shell Sort não são estáveis em suas formas típicas (embora existam modificações para torná-los estáveis, no caso de Quick e Selection, mas não é comum).

Um algoritmo de ordenação é estável quando dois elementos iguais (ou com chave de comparação igual) mantêm entre si a mesma ordem relativa antes e depois da ordenação. Ou seja, se temos elementos A e B com o mesmo valor e A aparece antes de B na lista original, após ordenar ambos continuarão na ordem A antes de B.

Exemplo de importância da estabilidade: Considere que você tenha uma lista de registros de clientes, onde cada registro tem nome e cidade. Suponha que inicialmente a lista esteja ordenada por ordem de cadastro (tempo em que cada cliente se registrou). Se quisermos ordenar essa lista por cidade (ordenar os clientes agrupados por cidade em ordem alfabética), um algoritmo estável garantirá que, para clientes da mesma cidade, a ordem original (por tempo de cadastro) seja preservada. Assim, após ordenar por cidade usando um algoritmo estável, os clientes estarão agrupados por cidade, mas dentro de cada cidade eles ainda estarão em ordem de cadastro. Se usarmos um algoritmo instável, essa ordem interna poderia se perder, e clientes da mesma cidade poderiam aparecer invertidos cronologicamente, o que talvez não seja desejado.

Outro exemplo: ordenar cartas de baralho por naipe mantendo a ordem por valor, ou vice-versa. Se primeiro ordenamos por valor e depois quisermos ordenar por naipe sem bagunçar a ordem de valores iguais, precisaríamos de uma ordenação estável.


7 Considere o vetor X = [7, 4, 5, 2]. Trace o passo a passo de como o Bubble Sort ordenaria X em ordem crescente, mostrando o estado do vetor após cada passagem completa pelo vetor.

Resolução: Vamos simular o Bubble Sort no vetor X:

Vetor inicial: [7, 4, 5, 2]

Passagem 1:

Compare 7 e 4 -> 7 > 4, então trocamos. Vetor: [4, 7, 5, 2]

Compare 7 e 5 -> 7 > 5, troca. Vetor: [4, 5, 7, 2]

Compare 7 e 2 -> 7 > 2, troca. Vetor: [4, 5, 2, 7]

Ao final da primeira passagem, o maior elemento (7) “borbulhou” para a última posição.
Vetor após 1ª passagem: [4, 5, 2, 7].

Passagem 2: (Agora não precisamos comparar o último elemento com nada, pois já está no lugar correto)

Compare 4 e 5 -> 4 < 5, nada a trocar. Vetor permanece [4, 5, 2, 7].

Compare 5 e 2 -> 5 > 2, troca. Vetor: [4, 2, 5, 7]

(Não comparamos 5 e 7, pois 7 já está fixo no fim nesta iteração ou podemos comparar mas não trocará)

Vetor após 2ª passagem: [4, 2, 5, 7]. (Note que 7 continua no fim, e agora 5 ficou na posição 2, mas 4 e 2 estão fora de ordem.)

Passagem 3:

Compare 4 e 2 -> 4 > 2, troca. Vetor: [2, 4, 5, 7].

Compare 4 e 5 -> 4 < 5, nada a trocar (e 5 e 7 também não trocam).

Vetor após 3ª passagem: [2, 4, 5, 7].

Passagem 4: Nesta passagem, percorreremos o vetor e não encontraremos nenhuma troca a ser feita, pois todos os pares adjacentes já estão em ordem (2<4, 4<5, 5<7). Quando uma passagem não realiza trocas, o algoritmo pode concluir que a lista está ordenada e finalizar (muitos códigos Bubble Sort implementam essa verificação). Após a 3ª passagem nosso vetor já estava ordenado, então a 4ª seria redundante.

O vetor final ordenado é [2, 4, 5, 7]. O passo a passo acima mostra o Bubble Sort movendo gradualmente os maiores elementos para o fim do vetor a cada passagem.


8 Se você precisa ordenar um vetor com milhões de elementos (um volume de dados muito grande), qual dos algoritmos estudados você usaria e por quê?

Resolução: Para ordenar um vetor muito grande (milhões de elementos), é preferível utilizar um algoritmo de complexidade O(n log n), como Quick Sort ou Merge Sort, ao invés dos algoritmos O(n^2) (Bubble, Selection, Insertion, Shell), pois estes últimos se tornariam impraticavelmente lentos com um número tão grande de elementos.

Entre Quick Sort e Merge Sort, a escolha pode depender de detalhes:

O Quick Sort costuma ser uma excelente opção em termos de performance prática média: ele é in-place (não requer memória extra significativa) e, escolhendo pivôs aleatoriamente ou via estratégia mediana-de-três, geralmente ordena milhões de itens com rapidez. Em implementações bem feitas, Quick Sort tende a superar Merge Sort em velocidade constante, devido à melhor localidade de referência e menos movimentação total em média.

O Merge Sort tem a vantagem de ter garantia de O(n log n) mesmo no pior caso e ser estável. No entanto, usa memória adicional O(n). Se a estabilidade for importante (por exemplo, preservar ordem de registros iguais) ou se estivermos lidando com listas encadeadas/disco externo, Merge Sort pode ser preferível.

Outra consideração: as linguagens de programação frequentemente oferecem métodos de ordenação otimizados. Por exemplo, o Python usa uma variação chamada Timsort, que combina ideias de Merge Sort e Insertion Sort, otimizada para vários cenários, inclusive dados quase ordenados. Timsort tem desempenho O(n \log n) no pior caso, é estável e muito eficiente. Em C++ e Java, implementações padrão costumam usar Quick Sort introspectivo (que cai para Heap Sort no pior caso para garantir O(n log n)) ou outros híbridos.

Se a pergunta espera uma resposta direta: eu usaria o Quick Sort (assumindo uma boa implementação) para milhões de elementos, porque ele é em média muito rápido (O(n \log n)) e utiliza memória in-place. Alternativamente, Merge Sort também seria uma escolha válida pelo desempenho garantido O(n \log n). O importante é evitar algoritmos quadráticos. Em suma, algoritmos de ordenação como Quick Sort ou Merge Sort (ou um híbrido moderno) são indicados para ordenar grandes volumes de dados, ao contrário de Bubble, Selection ou Insertion Sort, que seriam demasiadamente lentos em milhões de entradas.
