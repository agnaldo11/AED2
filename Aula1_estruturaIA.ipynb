1- Por que aprender Estrutura de Dados
R - Porque ajuda a organizar, armazenar e acessar dados de forma eficiente, melhorando o desempenho de programas e facilitando a resolução de problemas em programação e na Ciência de Dados.

2- O que é Estrutura de Dados
R - É uma forma de organizar e armazenar dados no computador para que possam ser utilizados de maneira rápida e eficiente.

3- Aplicações na área de Ciência de Dados
R - Na Ciência de Dados, estruturas de dados são usadas para:

3.1 armazenar grandes volumes de dados
3.2 organizar e processar informações
3.3 desenvolver modelos de Aprendizado de Máquina e análises de dados.
