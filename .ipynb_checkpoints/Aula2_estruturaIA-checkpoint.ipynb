1- O que significa Notação Big-O
R - A Notação Big‑O é uma forma de medir a eficiência de um algoritmo, mostrando quanto tempo ou memória ele precisa conforme a quantidade de dados aumenta.

2- Por que utilizar essa notação
R - Para comparar algoritmos e escolher o que tem melhor desempenho.

3- Tipos de complexidade
Respostas:

3.1- O(1) – constante
3.2- O(log n) – logarítmica
3.3- O(n) – linear
3.4- O(n²) – quadrática
3.5- O(2ⁿ) – exponencial
