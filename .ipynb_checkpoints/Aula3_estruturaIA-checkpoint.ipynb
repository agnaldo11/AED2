1- O que caracteriza um vetor não ordenado
R- Os elementos não estão organizados em ordem.

2- Passos para encontrar um valor

2.1- Começar no primeiro elemento.
2.2- Comparar com o valor procurado.
2.3- Se for diferente, verificar o próximo.
2.4- Repetir até encontrar ou chegar ao final.

3- Se o elemento estiver presente
R- A busca para e retorna a posição.

4- Se não estiver presente.
R- A busca termina e retorna não encontrado.

5- Eficiência
R- Complexidade O(n) na Notação Big-O.
