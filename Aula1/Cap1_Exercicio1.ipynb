Exercício 1: Crie uma lista em Python chamada frutas com os seguintes elementos: "maçã", "banana",
"laranja". Em seguida, adicione a fruta "morango" ao final da lista. Por fim, imprima a lista completa.

```python
# resolução exercicio 1
frutas = ['maça', 'banana', 'laranja']
frutas.append('morango')
print(frutas)

```

Exercício 2: Crie uma pilha (usando uma lista Python) chamada livros. Adicione os livros "O Pequeno
Príncipe", "Dom Quixote" e "1984" nesta ordem. Em seguida, simule a remoção do livro do topo da pilha
e imprima o livro removido e a pilha restante.

```python
#Resolução exercicio 2
livros = []
livros.append('O Pequeno Principe')
livros.append('Dom Quichote')
livros.append('1984')
livro_removido = livros.pop()
print(f'Livro removido: {livro_removido}')
print(f'Pilha restante: {livros}')
```
