1 - (ENADE) Em uma fila (estrutura de dados), é correto afrima que: 
    A. O último elemento que entra é o primeiro que sair.
    B. Não é possível inserir elementos enquanto realizamos remoções.
    C. A primeira operação de enfileiramento inseri no início da fila e a primeira de desenfileiramento remove do final. 
    D. O elemento removido numa operação dequeue é sempre o mais antigo (o primeiro a ter sido enfileirado).
    E. Filas são acessadas apenas pelo meio da estrutura. 

    Resposta: (D) (em uma fila, o ítem removido é justamente o elemento que esta na frente, ou seja, o mais antigo, seguindo FIFO).

4 - (Exercício Praticos) - Fila de impressão em um escritório. Implemente um sistema de gerenciamento de filas de impressão. Cada documento tem um nome. Documetos são processados na ordem em que chegam. Após a impressão de cada três documentos, a impressora precisa de uma pausa para "resfriamento", durante a qual nenhum novo documento pode ser processado por um ciclo. (mas novos documentos ainda podem ser adicionados a fila).

```python
def fila_impressao(documentos): 
    fila = queue()
    for doc in documentos: 
        fila.enqueue(doc)
    documentos_impressos = 0 
    ciclo = 0 
    while not fila.is_empty(): 
        #A cada 3 documentos impressos, a impressora precisa "esfriar"
        if documentos_impressos > 0 and documentos_impressos % 3 == 0: 
            print("Pausa para resfriaento da impressora.")
            ciclo += 1 
            continue # pula o ciclo sem imprimir
        doc_atual = fila.dequeue()
        print(f"Imprimindo documento: {doc_atual}")
        documentos_impressos += 1 
```

```python
# Exemplo de uso: 
fila_impressao(['Relatório.pdf', 'Contrato.docs', 'Imagem.png', 'Resumo.txt', 'Tabela.xlsx', 'Mapa.jpg'])
```
