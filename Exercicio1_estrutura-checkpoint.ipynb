```python
# Projeto 1- Agenda de contatos: 
contatos = {}
def adicionar_contato(nome, telefone, email):
    contatos[nome] = {'telefone': telefone, 'email': email}
    print(f'Contato {nome}, adicionado com sucesso') 

def buscar_contato(nome):
    if nome in contatos:
        print(f'Informações de {nome}:')
        print(f"Telefone: {contatos[nome]['telefone']}")
        print(f'Email: {contatos[nome]['email']}')
    else: 
        print(f'Contatos {nome} não encontrado')
def listar_contatos():
    print('\n lista de contatos:') 
    for nome,info in contatos.items():
        print(f'Nome: {nome}, Telefone: {info['telefone']}, E mail: {info['email']}') 
```

```python
# Exemplo uso: 
adicionar_contato('Ana', '997-345476', 'ana@email.com')
adicionar_contato('Bruno', '991-43568', 'bruno@email.com')
buscar_contato('Ana')
listar_contatos()
```

```python
# Projeto 2 - A lista de tarefas:
tarefa = ['Lavar a louça', 'Estudar para a prova', 'Fazer compras']
def adiocionar_tarefa(tarefa): 
    tarefas.append(tarefa)
    print(f"Tarefa '{tarefa}', adicionada.")
def marcar_concluida(tarefa):
    if tarefa in tarefa:
        tarefa.remove(tarefa)
        print(f"Tarefa '{tarefa}', marcada como concluida.")
    else:
        print(f'Tarefa {tarefa}, não encontrada na lista.')
def visualizar_tarefa():
    print('\nLista de tarefa:')
    for tarefa in tarefas:
        print(f'_{tarefa}')
        
```

```python
# Exemplo de uso: 
adicionar_tarefa('Pagar as contas')
marcar_concluida('Lavar a louça')
visualizar_tarefa()
```
