 1. (Teórica)

Em uma árvore binária, como são definidos os nós filho e pai? Qual alternativa está correta?

(A) Um nó que não tem galho para baixo é chamado de nó pai.

(B) O nó raiz é sempre uma folha.

(C) Cada nó filho em uma árvore binária tem exatamente dois pais.

(D) Um nó pai pode ter mais de dois filhos diretos.

(E) Se um nó não tem filhos, ele é chamado de nó folha. (x)

Resposta: Letra E. Em uma árvore, nós sem filhos são chamados folhas. A raiz é o nó mais alto (não um nó folha necessariamente), cada nó filho tem exatamente um único pai (não dois), e um nó pai em árvore binária pode ter até dois filhos (não sempre dois). Portanto, a alternativa E é a correta.

3. (Teórica)

Analise o trecho de código abaixo para percorrer uma árvore binária:

def percorre(no):
    if no:
        percorre(no.esquerda)
        percorre(no.direita)
        print(no.valor)

Esse percurso corresponde a qual método?

(A) Pré-ordem

(B) Em-ordem

(C) Pós-ordem (x)

(D) Largura (BFS)

(E) Nenhuma das anteriores

Resposta: Letra C. O código imprime `no.valor` após visitar recursivamente as subárvores esquerda e direita. Isso caracteriza o percurso em pós-ordem (esquerda → direita → nó). Em pré-ordem o print ocorreria antes das chamadas recursivas, e em in-ordem entre as duas chamadas. A alternativa (D) descreve busca em largura, que não usa recursão dessa forma. Portanto, trata-se de pós-ordem.


4. (Teórica)

Sobre altura e nível em uma árvore, assinale a opção correta:

(A) A raiz sempre está no nível 1.

(B) A altura de uma árvore vazia é 1.

(C) O nível de um nó é a quantidade de filhos que ele possui.

(D) A altura de uma árvore é o maior nível em que há um nó. (x)

(E) Em uma árvore, todos os nós têm a mesma altura.

Resposta: Letra D. Em estruturas de árvore, define-se nível como a distância em arestas (ou camadas) do nó até a raiz (ex.: raiz no nível 0, filhos no nível 1, etc.), e a altura da árvore é o maior nível ocupado por algum nó. Assim (D) está correto.

5 (Prática) Escreva uma função altura(raiz) em Python que recebe a raiz de uma árvore binária e retorna a altura da árvore (número de níveis). Dica: use recursão. Em seguida, teste-a na árvore criada no exemplo acima.

Solução: Podemos definir recursivamente que a altura de uma árvore vazia é 0, e caso contrário é 1 + o máximo entre as alturas das subárvores esquerda e direita. Por exemplo:

def altura(no):
    if no is None:
        return 0

    # calcula altura de cada subárvore
    alt_e = altura(no.esquerda)
    alt_d = altura(no.direita)

    return 1 + max(alt_e, alt_d)

Teste:
print(altura(raiz))  # usando a árvore do exemplo

Explicação: Se no é None, a árvore é vazia e a altura é 0. Caso contrário, calcula-se a altura das subárvores esquerda (alt_e) e direita (alt_d) recursivamente e retorna-se 1 (nível do nó atual) mais o maior destes valores. No nosso exemplo, a árvore tinha altura 3 (níveis 0–2).

6 (Prática) Implemente em Python uma função conta_folhas(raiz) que receba a raiz de uma árvore binária e retorne o número total de nós folha. Mostre a chamada da função no exemplo acima.

Solução: Podemos usar recursão: um nó é folha se não tiver filhos; a contagem total é a soma das contagens das subárvores esquerda e direita. Por exemplo:

def conta_folhas(no):
    if no is None:
        return 0
        
    # se no é folha
    if no.esquerda is None and no.direita is None:
        return 1

    # caso contrário, soma folhas das subárvores
    return conta_folhas(no.esquerda) + conta_folhas(no.direita)

Teste:
print(conta_folhas(raiz))  # com a árvore de exemplo

Explicação: A função verifica se o nó é None (árvore vazia) – então não há folhas. Se no não for None, verifica se ele não tem filhos à esquerda nem à direita; se sim, conta 1 (é uma folha). Senão, soma recursivamente as folhas de cada subárvore. No nosso exemplo a saída deverá ser 3 (os nós d, e e c).
